Regression: statmodel

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
model = ols('PRICE~RM+CRIM+PTRATIO',bos)
res = model.fit()
print(res.summary())
```

```python
from statsmodels.iolib.summary2 import summary_col

dfoutput = summary_col([res, res], stars=True)
print(dfoutput)
```

```python
beginningtex = """\\documentclass{report}
\\usepackage{booktabs}
\\begin{document}"""
endtex = "\end{document}"

f = open('summ.tex', 'w')
f.write(beginningtex)
f.write(dfoutput.as_latex())
f.write(endtex)
f.close()
```

Regression: sklearn

```python
from sklearn.linear_model import LinearRegression

X = bos.drop('PRICE', axis=1)
```

```python
lm = LinearRegression()
lm
```

```python
lm.fit(X, bos['PRICE'])
lm.score(X, bos['PRICE'])
```

```python
lm.coef_
```

```python
# coefficients
pd.DataFrame({'features':X.columns, 'estimatedCoefficients':lm.coef_})
```

```python
plt.hist(bos.PRICE, bins=25)
plt.hist(lm.predict(X), bins=25)
plt.legend(['orignal value', 'fitted value'])
plt.title('Predicted house price:$\hat{Y}_i$')
plt.xlabel('housing price')
```

```python
plt.scatter(bos.PRICE, lm.predict(X))
```

Not a straight line?

```python
# MSE
np.mean((bos.PRICE-lm.predict(X))**2)
```

## Training and testing


```python
import sklearn
```

```python
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(
    X, bos.PRICE, test_size=0.33, random_state = 0)
```

```python
X_train.shape
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)

pred_train = lm.predict(X_train)
pred_test = lm.predict(X_test)
```

```python
np.sum((y_train - pred_train)**2)
```

```python
print(lm.score(X_train, y_train))
print(lm.score(X_test, y_test))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# residual plots

plt.scatter(pred_train, pred_train-y_train, c='b', alpha=0.5)
plt.scatter(pred_test, pred_test-y_test, c='r', alpha=0.5)
plt.hlines(y=0, xmin=0,xmax=50)
plt.xlabel('prediction')
plt.ylabel('residuals')
```
